```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data =  pd.read_csv("/content/sample_data/california_housing_test.csv")
print(data)
sns.lineplot(x="total_rooms",y="total_bedrooms",data=data)
plt.show()
```

```python
# bar plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("titanic")
print(data)
sns.barplot(x="sex",y="age",data=data,hue = "class")
plt.show()
# to remove error bars
sns.barplot(x="sex",y="age",data=data,hue = "class",errorbar=None)
plt.show()
```

```python
# histogram
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("titanic")
print(data)
sns.histplot(data=data,x="age",hue="who",kde=True)
plt.show()
```

```python
# scatter plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
print(data)
sns.scatterplot(x="total_bill",y="tip",data=data,hue="day",size="size")
plt.show()
```

```python
# heatmap
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
print(data)
gp = data.groupby("day").agg({"tip":"mean"})
sns.heatmap(gp)
plt.show()
```

```python
# count plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("titanic")
print(data)
sns.countplot(x="sex",data=data,hue="class")
plt.show()
```

```python
# violin plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
print(data)
sns.violinplot(data=data,x="total_bill")
plt.show()
```

```python
# pair plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
print(data)
sns.pairplot(data)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
print(data)
sns.pairplot(data,hue="species")
plt.show()
```

```python
# strip plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
sns.stripplot(x="day",y="total_bill",data=data,hue="sex",dodge = True,jitter=0.3)
plt.show()
```

```python
# box plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
print(data)
sns.boxplot(x="day",y="total_bill",data=data,hue="sex")
plt.show()
```

```python
# categorical plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
print(data)
sns.catplot(x="day",y="total_bill",data=data,hue="sex",kind="bar")
```

```python
# style and colors in plots

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = sns.load_dataset("titanic")
print(data)
sns.set_style(style="darkgrid")
sns.set_palette("viridis")
sns.barplot(data=data,x="sex", y="age")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.palplot(sns.color_palette("viridis",10))
plt.show()
```

```python
# relational plot
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
print(data)

sns.relplot(data=data,x="tip",y="total_bill",hue="sex",col="day")
plt.show()
```

```python
# swarm plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
print(data)

sns.swarmplot(data=data,x="day",y="total_bill" , hue ="sex",dodge=True)
plt.show()
```

```python
# kde plot (kernel density estimation)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
print(data)

sns.kdeplot(data=data,x="total_bill",hue="sex",multiple="stack")
plt.show()
```
