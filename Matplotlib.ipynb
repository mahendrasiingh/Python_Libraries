```python
# Bar Chart
import matplotlib.pyplot as plt
x = [50,60,90,70,10]
y = ["part1","part2","part3","part4","part5"]
color = ["red","green","blue","yellow","pink"]
plt.bar(x,y,color = color,edgecolor = "black")
plt.bar(x, y, color=color, edgecolor="black", width=8)  # Adjust 0.5 as needed
plt.xlabel("marks",fontsize = 15)
plt.ylabel("parts",fontsize = 15)
plt.title("bar chart",fontsize = 15)
plt.show()
```

```python
# Line Plot
import matplotlib.pyplot as plt
y = [500,600,100,700,100]
x = ["day1","day2","day3","day4","day5"]
y1 = [100,200,400,300,500]
plt.plot(x,y,marker = "d",ls = "--",label = "week1",alpha = 0.5)
plt.plot(x,y1,marker = "^",label = "week2")
plt.xlabel("days")
plt.ylabel("sales")
plt.legend(["week1","week2"])
plt.show()
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(1,10,50)
y = np.random.randint(10,100,50)
color = np.random.randint(10,100,50)
sizes = np.random.randint(10, 200, 50)
plt.scatter(x,y,marker="*",cmap="hot",c=color,s=sizes)
plt.colorbar()
plt.show()

```

```python
# Pie Chart
import matplotlib.pyplot as plt
brands = ["acer","hp","lenevo","apple","samsung","asus"]
x = [10,35,12,30,18,10]
ex = [0,0,0,0,0,0.1]
plt.pie(x, labels=brands,explode=ex,shadow=True,autopct="%.2f",startangle=45)
plt.show()
```

```python
# Box Plot
import matplotlib.pyplot as plt
import numpy as np
l = [10,20,30,40,88,75,30,20,46,50,60,70,80,90,100]
plt.boxplot(l)
plt.show()
```

```python
# Histogram
import matplotlib.pyplot as plt
x = [10,20,30,40,88,75,30,20,46,50,60,70,80,90,100]
plt.hist(x,bins=15,edgecolor = "black",color="magenta")
plt.show()
```

```python
# Violin Plot
import matplotlib.pyplot as plt
x = [10,88,75,30,20,46,50,60,70,80,90,100]
plt.violinplot(x,showmedians=True)
plt.show()
```

```python
# Stem plot
import matplotlib.pyplot as plt
x = [10,20,30,40,88,75,30,20,46,50,60,70,80,90,100]
plt.stem(x,linefmt="--",markerfmt="d")
plt.show()
```

```python
# Stack plot
import matplotlib.pyplot as plt
days = [1,2,3,4,5,6,7]
nop1 = [5,10,30,20,35,50,60]
nop2 = [20,40,50,10,80,55,70]
nop3 = [8,30,25,60,70,90,100]
plt.stackplot(days,nop1,nop2,nop3,labels=["week1","week2","week3"])
plt.legend()
plt.show()
```

```python
# Step plot
import matplotlib.pyplot as plt
x = ["day1","day2","day3","day4","day5"]
y = [30,40,25,30,40]
plt.step(x,y)
plt.show()
```

```python
# Legend

import matplotlib.pyplot as plt
x = [10,20,30,40,50,60]
y = [45,60,70,30,20,10]
y1 = [41,60,50,20,30,70]
plt.figure(figsize=[5,3])
plt.plot(x,y , label = "male")
plt.plot(x,y1, label = "female")
plt.legend()
plt.show()
```

```python
# Sub plot
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [45,34,56,23,45]
plt.subplot(2,2,1) # row,column,chart number
plt.bar(x,y)

x = [1,2,3,4,5]
y = [45,34,56,23,45]
plt.subplot(2,2,2)
plt.plot(x,y)

x = [5,6,7,8,9]
y = [20,50,25,10,60]
plt.subplot(2,2,3)
plt.plot(x,y)

x = [5,6,7,8,9]
y = [20,50,25,10,60]
plt.subplot(2,2,4)
plt.plot(x,y)
plt.suptitle("subplot")
plt.show()
```
